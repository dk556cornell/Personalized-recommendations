```python
import pandas as pd
from faker import Faker
import numpy as np
import random
```

```python
fake = Faker()
```

```python
num_customers = 500
num_products = 50
num_purchases = 5000
```

```python
customers = [f'C{i:03d}' for i in range(1, num_customers + 1)]
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Books', 'Grocery', 'Health & Beauty']
products = {}
#Each unique product ID has a set price that is within an approriate range by category
for i in range(1, num_products + 1):
    category = random.choice(categories)
    # Price ranges by category
    if category == 'Electronics':
        price_base = random.uniform(100, 1000)
    elif category in ['Home & Kitchen', 'Health & Beauty', 'Clothing']:
        price_base = random.uniform(30, 200)
    elif category == 'Groceries':
        price_base = random.uniform(20, 100)
    else:
        price_base = random.uniform(10, 50)
    products[f'P{i:03d}'] = (category, round(price_base, 2))
```

```python
purchase_data = []
for _ in range(num_purchases):
    customer_id = random.choice(customers)
    product_id = random.choice(list(products.keys()))
    category, amount = products[product_id]
    date = fake.date_between(start_date='-1y', end_date='today')
    purchase_data.append([customer_id, product_id, category, amount, date])
```

```python
df = pd.DataFrame(purchase_data, columns=['Customer ID', 'Product ID', 'Product Category', 'Purchase Amount', 'Purchase Date'])
df.head()
```

```python
df.to_csv('purchase_data.csv', index=False)
```
